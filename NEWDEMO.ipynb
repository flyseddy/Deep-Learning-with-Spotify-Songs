```python
import matplotlib.pyplot as plt
import math
fig = plt.figure(figsize=(12,8))

# Normalize these values Min Max Normalization
new_df = df.loc[:,'danceability':'liked']
new_df = (new_df - new_df.min()) / (new_df.max() - new_df.min())
liked_df = new_df[new_df['liked'] == 1]
liked_df.drop('liked', axis=1, inplace=True)

categories = list(liked_df.columns)

N = len(categories)

value = list(liked_df.mean())

value += value[:1]

angles = [n/float(N) * 2*math.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, value)
plt.fill(angles, value, alpha=0.3)

plt.xticks(angles[:-1],categories, size=15)
plt.yticks(color='grey', size=15)
plt.title('Liked Songs')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
fig = plt.figure(figsize=(12,8))

disliked_df = new_df[new_df['liked'] == 0]
disliked_df.drop('liked', axis=1, inplace=True)

categories = list(disliked_df.columns)

N = len(categories)

value = list(disliked_df.mean())

value += value[:1]

angles = [n/float(N) * 2*math.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, value)
plt.fill(angles, value, alpha=0.3)

plt.xticks(angles[:-1],categories, size=15)
plt.yticks(color='grey', size=15)
plt.title('Disliked Songs')
plt.show()
```

```python
# Libraries for preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
X = songs
y = songs['liked'] # Target (Liked or Not Liked)
```

```python
X
```

```python
# Split train, validation, test data with 60/20/20 split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.25, random_state=1) # .25 x .8 = .20 - 20% validation, 60% train, 20 % test
```

```python
X_train.head()
```

```python
X_train = X_train.drop(['name', 'artist'], axis=1)
```

```python
X_train = X_train.drop(['liked'], axis=1)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_Test_Copy = X_test[:]
```

```python
X_test = X_test.drop(['name', 'artist', 'liked'], axis=1)
```

```python
X_test.head()
```

```python
Y_train.head()
```

```python
Y_test.head()
```

```python
X_val.head()
```

```python
X_val = X_val.drop(['name', 'artist', 'liked'], axis=1)
```

```python
Y_val.head()
```

```python
# Normalize X_train, X_Val, X_Test
# Min Max (0-1 scale)
X_train = MinMaxScaler().fit_transform(X_train)
X_val = MinMaxScaler().fit_transform(X_val)
X_test = MinMaxScaler().fit_transform(X_test)
```

```python
import numpy as np
# Vectorize our labels
Y_train = np.asarray(Y_train).astype('float32')
Y_test = np.asarray(Y_test).astype('float32')
Y_val = np.asarray(Y_val).astype('float32')
```

```python
Y_val
```

```python
Y_test
```

```python
# Let's build our model!
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
      layers.Dense(128, activation='relu', input_shape=[13]),
      layers.Dense(128, activation='relu'),
      layers.Dense(1, activation='sigmoid'),
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy']
)
history = model.fit(X_train, Y_train, epochs=8, batch_size=64, validation_data=(X_val, Y_val))
```

```python
history_dict = history.history
history_dict.keys()
```

```python
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
acc_values = history_dict['binary_accuracy']
val_acc_values = history_dict['val_binary_accuracy']

epochs = range(1, len(acc_values) + 1)
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.clf()
plt.plot(epochs, acc_values, 'bo', label='Training acc')
plt.plot(epochs, val_acc_values, 'b', label='Validation acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
model.save("SpotifyPrediction.h5")
```

```python
results = model.evaluate(X_test, Y_test) # 77% Accuracy on Test Set
```
